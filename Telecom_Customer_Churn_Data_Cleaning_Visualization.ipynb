```python
# Shape and data types
print("Dataset Shape:", df.shape)
display(df.head())

df.info()
```

```python
# Missing values
missing = df.isnull().sum().sort_values(ascending=False)
missing[missing > 0]
```

```python
# Duplicate records
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df['Churn Label'] = df['Churn Label'].map({
    'Yes': 1,
    'No': 0
})
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])

for col in df.select_dtypes(include=['int64','float64']).columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Churn Label', data=df)
plt.title('Churn Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Monthly Charges'], kde=True)
plt.title('Distribution of Monthly Charges')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Tenure Months'], kde=True)
plt.title('Customer Tenure Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    x='Contract',
    hue='Churn Label',
    data=df
)
plt.xticks(rotation=20)
plt.title('Contract Type vs Churn')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    x='Contract',
    hue='Churn Label',
    data=df
)
plt.xticks(rotation=20)
plt.title('Contract Type vs Churn')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(
    x='Payment Method',
    hue='Churn Label',
    data=df
)
plt.xticks(rotation=45)
plt.title('Payment Method vs Churn')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    x='Internet Service',
    hue='Churn Label',
    data=df
)
plt.title('Internet Service vs Churn')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(12,8))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y=df['Monthly Charges'])
plt.title('Outlier Detection: Monthly Charges')
plt.show()
```

```python
churn_rate = df.groupby('Contract')['Churn Label'].mean() * 100

churn_rate.plot(kind='bar', figsize=(8,5))
plt.ylabel("Churn Rate (%)")
plt.title("Churn Rate by Contract Type")
plt.show()
```
